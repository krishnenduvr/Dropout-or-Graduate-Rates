# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
d=c.drop(["Target"],axis="columns")
d
```

```python
e=c.Target
e
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(d,e,test_size=0.9)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
import warnings
warnings.filterwarnings("ignore")
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

# CONFUSIN_MATRIX OF XGBCLASSIFIER

# VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr, kendalltau
```

```python
target_counts = a['Target'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral', 'cyan'])
plt.title('Graduationa status distributation')
plt.show()
```

```python
x=a.drop(['Gender', 'Target'], axis=1)
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(x.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=a, x='Target', hue='Course', palette='viridis')

plt.title('Target VS Course')
plt.xlabel('Target')
plt.ylabel('Course')
plt.show()
```
